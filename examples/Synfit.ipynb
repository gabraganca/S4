# Synfit

`Synfit` is a rotuine that find the a set of given parameters that best fit a spectral line or region. It allows the user to set any number of parameters. For this set of parameters, `Synfit` synthesize spectra using all combinations possible and find the best value by minimizing the $\chi^2$.


The first version was developed by Dr. I. Hubeny on IDL. 

On the examples below, we will automatically fit the O II line at 4144 &#x212b;. The first test is on the line synthetized using Synfit, and the second one is with the line observed from the star HD 74575.

```python
import matplotlib.pyplot as plt
%matplotlib inline

import s4
from s4.synthesis import Synplot
from s4.fitting import Synfit
```

## Synthetic spectrum

First, we synthesize the spectral line.

```python
## Creates a fake spectrum
params = dict(teff=23000, logg=3.6, vrot=15, vmac_rt=10,
              wstart=4405, wend=4425, noplot=True, relative=1,
              abund='[8,8,8.69]', observ='test_spectrum.dat')


syn = Synplot(params['teff'], params['logg'],
              wstart=params['wstart'], wend=params['wend'],
              relative=params['relative'], vrot=params['vrot'],
              abund=params['abund'], vmac_rt=params['vmac_rt'])

syn.savetxt(params['observ'])
syn.plot(ymin=0.5, windows=[4413.8, 4416.1])
```

We can test with just one parameter, vsini.

`Synfit` needs an argument which is a dictionary containing the parameters to be fitted as the keys. The values to be fitted should be passed as a list with three values indices: first is the lower value, the second is the uppermost value and the thrid, the step.

Also, `Synfit` accpets other arguments containing the parameters that are fix and other non-stellar parameters, as the path to the obseravtion spectrum. For more details on these parameters, please refer to the `Synplot` documentation.

We then create the `Synfit` instance and ask it to fit.

```python
## Test with one parameter
params_copy = params.copy()
del params_copy['vrot']

fit = Synfit({'vrot':[0, 30, 5]},
             **params_copy)
fit.fit()
```

The best fit values are stored in the `best_fit` variable.

```python
for key, val in fit.best_fit.iteritems():
    print '{} = {}'.format(key, val)
```

We can plot the best fit.

```python
fit.best_plot(title='default')
```

And we can can analyse the $\chi^2$ with the parameter.

```python
fit.plot_chisquare()
```

We can also set any numbers of parameters, but let's stay with two: vsini and vmac_rt.

```python
## Test with two parameter
params_copy = params.copy()
del params_copy['vrot']

fit = Synfit({'vrot':[0, 30, 5], 'vmac_rt':[0, 30, 5]},
             **params_copy)
fit.fit()
```

And the values found are:

```python
for key, val in fit.best_fit.iteritems():
    print '{} = {}'.format(key, val)
```

We then plot the fit.

```python
fit.best_plot(title='default')
```

And the $\chi^2$ values related to the parameters are:

```python
# Linear interpolation
fit.plot_chisquare('linear')
plt.title('Linear interpolation')

# Cubic Interpolation
fit.plot_chisquare('cubic')
_ = plt.title('cubic interpolation')
```

If one of the parameter is the abundance of a chemical element, the user only needs to use its chemical symbol.

On the example below, we are fitting the above spectral line to find the best values of the abundance of Oxygen (`O`).

```python
## Test with abundance
params_copy = params.copy()
del params_copy['abund']

fit = Synfit({'O':[8.60, 8.75, 0.01]},
             **params_copy)
fit.fit()
```

And the values found are:

```python
for key, val in fit.best_fit.iteritems():
    print '{} = {}'.format(key, val)
```

And the best fit on the plot is shown below.

```python
fit.best_plot(title='default')
```

## Real spectral line

We can now try the same exercise with a real spectral line. We keep showing the O II line at 4144 Å, but now for a real star: HD 74575.

```python
hd74575 = s4.io.load_spectrum('../notebooks/HD74575_f09601_4200_4500_norm.fits')

fig = plt.figure(figsize=(15,10))

ax=fig.add_subplot(111)

ax.plot(hd74575[:,0], hd74575[:,1], 'r-')

ax.set_xlim([4405, 4425])
ax.set_ylim([0.5, 1.05])

ax.set_title('HD 74575: O II lines at 4414 $\AA$')

s4.plottools.plot_windows([4414, 4416.4])
```

Let's repeat the exercise. First with just 

```python
params = dict(teff=23000, logg=3.6, wstart=4405, wend=4425, rv=15, relative=1, scale=1, vturb=10, vrot=10, vmac_rt=20,
              observ='../notebooks/HD74575_f09601_4200_4500_norm.fits', noplot=True, windows=[4413.8, 4416.1])
```

We then create the `Synfit` instance and ask it to fit.

```python
fit = s4.fitting.Synfit( {'vrot':[0, 50, 5]}, 
                        **params)
fit.fit()
```

The best fit values are stored in the `best_fit` variable.

```python
for key, val in fit.best_fit.iteritems():
    print '{} = {}'.format(key, val)
```

And it is possible to plot the best fit.

```python
fit.best_plot(title='default')
```

The $\chi^2$ Vs. vsini is shown below:

```python
fit.plot_chisquare()
```

We try again two parameters, vsini and vmac_rt:

```python
fit = s4.fitting.Synfit( {'vrot':[0, 50, 5], 'vmac_rt':[0,50,5]}, 
                        **params)
fit.fit()
```

The values fitted are shown below.

```python
for key, val in fit.best_fit.iteritems():
    print '{} = {}'.format(key, val)
```

The plot with the best fit.

```python
fit.best_plot(title='default')
```

And the value $\chi^2$ according to the parameters can be seen on the plot below.

```python
# Linear interpolation
fit.plot_chisquare('linear')
plt.title('Linear interpolation')

# Cubic Interpolation
fit.plot_chisquare('cubic')
_ = plt.title('cubic interpolation')
```

And we can also fit abundance for an observed spectral line.

```python
fit = s4.fitting.Synfit( {'O':[8.70, 8.90, 0.01]}, 
                        **params)
fit.fit()
```

The values fitted are:

```python
for key, val in fit.best_fit.iteritems():
    print '{} = {}'.format(key, val)
```

The plot with the best fit.
